## Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge amount must be at least â‚¹50")
        elif amount not in valid_plans:
            print("Invalid plan selected")
        else:
            print("Recharge Successful")
            break

validate_recharge()
```

## Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, ": Reorder Alert")
        else:
            print(product, ": Stock OK")

products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(products)
```

## Problem 3: Student Result Processing System

```python
def check_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks_list = [60, 45, 70, 55]
print(check_result(marks_list))
```

## Problem 4: Cab Fare Estimator with Retry

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (distance * 12)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare

while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(km, peak_hour)
    print("Total Fare:", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

## Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]
print(check_attendance(attendance_list))
```

## Problem 6: Password Strength Checker

```python
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password(pwd))
```
