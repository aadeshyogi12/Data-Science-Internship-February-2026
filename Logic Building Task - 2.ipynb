## Problem 1: Unique Words in a Sentence

```python
sentence = input("Enter sentence: ")

words = sentence.split()

unique_words = []

for word in words:
    if word not in unique_words:
        unique_words.append(word)

print("Unique words count:", len(unique_words))
print("Unique words:", set(unique_words))
```

## Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_salary = 0
highest_employee = ""

for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

## Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

## Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

## Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0
total_days = len(attendance)

for day in attendance:
    if day == "P":
        present_count += 1

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

## Problem 7: Count Character Frequency

```python
text = "python"

char_count = {}

for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1

print(char_count)
```

## Problem 9: Check if a Key Exists in a Dictionary

## Problem 10: Calculate Average Marks

```python
marks = [65, 70, 80, 75]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print("Average Marks:", average)
```
