## Task 1: Conceptual Understanding (Mandatory – Written)

#### 1. What is the difference between "Love" and "love" in NLP?

+ In NLP, text is case-sensitive by default, which means "Love" and "love" are treated as completely different words. This can mess things up because it increases the number of unique words in the dataset and makes the model less efficient. To avoid this, we usually convert everything to lowercase during preprocessing so both "Love" and "love" are treated the same.

#### 2. What happens if stopwords are not removed?

+ If we don't remove stopwords, the dataset ends up full of common words like "is", "the", and "and" that don't really add much meaning. This makes the data more bulky and noisy. Due to which, models take more time to process the data and may not perform as well since they're focusing on useless words instead of meaningful ones.

#### 3. Provide two real-world scenarios where removing stopwords can be harmful.

+ In sentiment analysis, removing words like "not" can completely change the meaning. For example, "not good" would become just "good". This is obviously wrong!

+ In question-answer systems, removing words such as who, what or where takes away the main intent of the question making it harder for a model to understand what's being asked.

#### 4. Explain the difference between stemming and lemmatization.

+ Stemming: A rule-based approach where we remove prefixes or suffixes to get to some base form of a word without considering its context.

+ Examples: "running" → "run", "studies" → "studi" (may not be a real word) Lemmatization: An advanced process that takes into account the context of a + word before converting it into its actual dictionary base form so that the output is always a proper word.

+ Examples: "running" → "run", "studies" → "study", "better" → "good"

## Task 2:Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove repeated characters (soooo → so)
    text = re.sub(r'(.)\1+', r'\1', text)

    # Remove special characters (keep words)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short tokens (except no, not)
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence
```

## Task 3: Stress Testing

```python
#test cases
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in sentences:

    #preprocessing
    tokens, clean = preprocess_text(s)

    print("Original Text:", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print()

    results.append((s, tokens, clean))
```

## Task 4: Token Analytics

```python
for s, tokens, _ in results:

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    # avoiding division by zero if no tokens
    if total_tokens == 0:
        avg_length = 0
    else:
        avg_length = sum(len(w) for w in tokens) / total_tokens

    print("Sentence:", s)
    print("Total tokens:", total_tokens)
    print("Unique tokens:", unique_tokens)
    print("Average token length:", round(avg_length, 2))
    print()
```

+ Most meaningful tokens retained:

## Task 5: Frequency Analysis

```python
from collections import Counter

all_tokens = []

# combining all tokens from all sentences
for _, tokens, _ in results:
    all_tokens.extend(tokens)

freq = Counter(all_tokens)

print("Top 10 most frequent words:")
print(freq.most_common(10))

print("\nTop 5 least frequent words:")
print(freq.most_common()[-5:])
```

## Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```
