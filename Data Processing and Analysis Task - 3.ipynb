## Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

max_score = max(employees.values())
top = []

for name in employees:
    if employees[name] == max_score:
        top.append(name)

print("Top Performers Eligible for Bonus:", ", ".join(top), "(Score:", max_score, ")")
```

## Problem Statement 2: Search Query Keyword Analysis

```python
import string

text = "Buy mobile phone buy phone online"
text = text.lower()
text = text.translate(str.maketrans("", "", string.punctuation))

words = text.split()
count = {}

for w in words:
    if w in count:
        count[w] += 1
    else:
        count[w] = 1

result = {}

for key in count:
    if count[key] > 1:
        result[key] = count[key]

print(result)
```

## Problem Statement 3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid)
```

## Problem Statement 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain = {}
total = len(emails)

for mail in emails:
    d = mail.split("@")[1]
    if d in domain:
        domain[d] += 1
    else:
        domain[d] = 1

for d in domain:
    percent = (domain[d] / total) * 100
    print(d + ":", str(int(percent)) + "%")
```

## Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg = sum(sales) / len(sales)

for i in range(len(sales)):
    if sales[i] >= avg * 1.29:
        print("Day", i + 1, ":", sales[i])
```

## Problem Statement 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

freq = {}

for u in user_ids:
    if u in freq:
        freq[u] += 1
    else:
        freq[u] = 1

for u in freq:
    if freq[u] > 1:
        print(u, "→", str(freq[u]), "times")
```
